```python
# Add the Pandas dependency.
import pandas as pd
import numpy as np
```

```python
url_co2 = 'https://raw.githubusercontent.com/Ricardolpz99/Turn-Green/main/Database/co-emissions-per-capita.csv'
CO2_emissions_df = pd.read_csv(url_co2)
```

```python
url_RE_gen = 'https://raw.githubusercontent.com/Ricardolpz99/Turn-Green/main/Database/renewable-energy-gen.csv'
RE_gen_df=pd.read_csv(url_RE_gen)
```

```python
CO2_emissions_df.head(10)
```

```python
RE_gen_df.head(10)
```

```python
Climate_data_complete_generated_df = pd.merge(CO2_emissions_df, RE_gen_df)
Climate_data_complete_generated_df.head(60)
```

```python
 len(Climate_data_complete_generated_df["Entity"].unique())
```

```python
 len(Climate_data_complete_generated_df["Code"].unique())
```

```python
Climate_data_complete_generated_df.loc[Climate_data_complete_generated_df.Entity == 'Africa', 'Code'] = "AFR"
```

```python
Climate_data_complete_generated_df.loc[Climate_data_complete_generated_df.Entity == 'Europe', 'Code'] = "EUR"
```

```python
Climate_data_complete_generated_df.loc[Climate_data_complete_generated_df.Entity == 'European Union (27)', 'Code'] = "EURU"
```

```python
Climate_data_complete_generated_df.drop(Climate_data_complete_generated_df.index[Climate_data_complete_generated_df['Solar Generation - TWh'] == 0], inplace=True)
```

```python
Climate_data_complete_generated_df.drop(Climate_data_complete_generated_df.index[Climate_data_complete_generated_df['Hydro Generation - TWh'] == 0], inplace=True)
```

```python
Climate_data_complete_generated_df.drop(Climate_data_complete_generated_df.index[Climate_data_complete_generated_df['Wind Generation - TWh'] == 0], inplace=True)
```

```python
Climate_data_complete_generated_df.drop(Climate_data_complete_generated_df.index[Climate_data_complete_generated_df['Geo Biomass Other - TWh'] == 0], inplace=True)
```

```python
Climate_data_complete_generated_df.to_csv('./Database/Clean/Climate_data_complete_generated.csv')
```
